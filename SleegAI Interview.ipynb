### Final: Detection of dock shortage

```python
def dock_shortage(df, station_id):
    out = []
    count = 0
    for ids in station_id:
        from_station_info = df[df["from_station_id"] == ids].describe()["tripduration"][:2]
        to_station_info = df[df["to_station_id"] == ids].describe()["tripduration"][:2]
        if ((to_station_info[0] < from_station_info[0]) & (from_station_info[1] < to_station_info[1])):
            out.append(ids)
#         count += 1    
#         print(count)
    return out

shortage_list = dock_shortage(q4_trip_df, station_df["id"])
```

```python
print("There are in total {} number of dock station are in shortage.".format(len(shortage_list)))
```

### Final: Enough Dock Station Detection

```python
def dock_enough(df, station_id):
    out = []
    count = 0
    for ids in station_id:
        from_station_info = df[df["from_station_id"] == ids].describe()["tripduration"][:2]
        to_station_info = df[df["to_station_id"] == ids].describe()["tripduration"][:2]
        if ((to_station_info[0] > from_station_info[0]) & (from_station_info[1] > to_station_info[1])):
            out.append(ids)
#         count += 1    
#         print(count)
    return out

enough_list = dock_enough(q4_trip_df, station_df["id"])
```
